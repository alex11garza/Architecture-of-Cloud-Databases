# Inner Join Queries

```python
#INNER JOIN pagila.country and pagila.city by using country_id which is present on both tables WHERE country.country_id = 44 or = 103 ORDERED BY city with a LIMIT of 10
!psql -c "SELECT country.country, city.city, country.country_id FROM pagila.country INNER JOIN pagila.city on city.country_id = country.country_id WHERE country.country_id = 44 OR country.country_id = 103 ORDER BY city LIMIT 10"
```

```python
#INNER JOIN film id from pagila.film and pagila.film_actor ORDERED BY title LIMTI of 10
!psql -c "SELECT f.rating, f.title, fa.film_id, fa.actor_id FROM pagila.film_actor fa INNER JOIN pagila.film f ON fa.film_id = f.film_id WHERE fa.actor_id = 30 ORDER BY f.title LIMIT 10"
```

```python
#INNER JOIN store_id between pagila.store table and pagila.staff table  ORDER BY manager_staff LIMIT to 2
!psql -c "SELECT store.address_id, store.store_id FROM pagila.store INNER JOIN pagila.staff on store.store_id = staff.store_id ORDER BY store.manager_staff_id LIMIT 10"
```

```python
#INNER JOIN customer_id from pagila.payment table and pagila.customer table WHERE p.amount = $0.99 ORDER BY customer.last_name, customer.first_name, and payment.amount LIMIT to 10
!psql -c "SELECT c.last_name, c.first_name, p.amount FROM pagila.payment p INNER JOIN pagila.customer c ON p.customer_id = c.customer_id WHERE p.amount = 0.99 ORDER BY c.last_name, c.first_name, p.amount LIMIT 10"
```

```python
#INNER JOIN between pagila.store and pagila.staff FROM manager_staff_name_id ORDER BY last_name, first_name, and store_id
!psql -c "SELECT sf.last_name, sf.first_name, sr.store_id FROM pagila.store sr INNER JOIN pagila.staff sf ON sr.manager_staff_id = sf.staff_id ORDER BY sf.last_name, sf.first_name, sr.store_id"
```

# Outer Join Queries

```python
#OUTER JOIN pagila.actor and pagila.film_actor WHERE film_id = 1 ORDER BY film_id LIMIT 10
!psql -c "SELECT actor.first_name, actor.last_name, actor.actor_id, film_actor.film_id FROM pagila.actor FULL OUTER JOIN pagila.film_actor on actor.actor_id = film_actor.actor_id WHERE film_actor.film_id = 1 ORDER BY film_actor.film_id LIMIT 10"
```

```python
#OUTER JOIN pagila.country and pagila.city by country.city_id = city.city_id WHERE country.country_id , 50 ORDER BY country.country LIMIT 10|
!psql -c "SELECT cty.country,cy.city,cty.country_id FROM pagila.address a LEFT OUTER JOIN pagila.city cy ON a.city_id = cy.city_id LEFT OUTER JOIN pagila.country cty ON cy.country_id = cty.country_id WHERE cty.country_id < 50 ORDER BY cty.country,cy.city LIMIT 10"
```

```python
#OUTER JOIN pagila.category and pagila.film by category.category_id and film.category_id WHERE category.name = 'Horror' ORDER BY film.title LIMIT 10
!psql -c "SELECT f.title, c.name FROM pagila.film_category fc LEFT OUTER JOIN pagila.film f ON fc.film_id = f.film_id LEFT OUTER JOIN pagila.category c ON fc.category_id = c.category_id WHERE c.name = 'Horror' or c.name = 'Documentary' ORDER BY f.title LIMIT 10"
```

# Self Join Queries

```python
# SELF-JOIN actor table is compared with itself to extract people who have the same last name but different actor ids
!psql -c "SELECT a.actor_id, a.first_name, a.last_name, b.actor_id, b.first_name, b.last_name FROM pagila.actor a INNER JOIN pagila.actor b ON a.actor_id != b.actor_id AND a.last_name = b.last_name ORDER BY a.last_name, a.first_name LIMIT 10"
```
